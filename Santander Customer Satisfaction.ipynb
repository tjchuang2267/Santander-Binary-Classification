# Dimensionality Reduction
#### We have a huge number of features, so we want to use PCA to reduce dimensionality of our dataset

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
y = train.TARGET
X = train.drop(['TARGET'], axis = 1)
```

```python
# Normalizing our data
scaler = StandardScaler()

scaler.fit(X)
scaler.fit(test)

X_scaled = scaler.transform(X)
test_scaled = scaler.transform(test)
```

```python
X_scaled
```

```python
pca = PCA(n_components = 3)
```

```python
pca.fit(X_scaled)
```

```python
pca_train_set = pca.transform(X_scaled)
pca_test = pca.transform(test_scaled)
```

```python
print('PCA Training Shape:', pca_train_set.shape)
print('PCA Test Shape:', pca_test.shape)
```

## Creating a Scree Plot

```python
PC_values = np.arange(pca.n_components_) + 1

plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth = 2, color = 'blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Train Test split to check model performance
X_train, X_test, y_train, y_test = train_test_split(pca_train_set, y, test_size = 0.3)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

### Confusion Matrix

```python
from sklearn import metrics
```

```python
y_pred = model.predict(X_test)
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])

cm_display.plot()
plt.show
```

#### Our model has a 96% training accuracy, we are now ready to apply the model to the test dataset
